```python
import pandas as pd



# Step 1: Find the median age
median_age = data['Age'].median()

# Step 2: Fill missing values with the median
data['Age'] = data['Age'].fillna(median_age)
print(data['Age'])
# Now the 'Age' column has missing values replaced with the median age
```

```python
import pandas as pd

# Assuming your DataFrame is called 'data'

# Step 1: Calculate missing value percentages
missing_value_percentages = (data.isnull().sum() / len(data)) * 100

# Step 2: Find columns with high missing values
columns_with_high_missing = missing_value_percentages[missing_value_percentages > 50].index

# Step 3: Print column names (if any)
if any(missing_value_percentages > 50):
    print("Columns with more than 50% missing values:")
    print(columns_with_high_missing.tolist())
else:
    print("No columns have more than 50% missing values.")

# Step 4: Filter DataFrame (optional, improved)
if any(missing_value_percentages > 50):
    low_missing_columns = missing_value_percentages[missing_value_percentages <= 50].index  # Get column names
    df_filtered = data[low_missing_columns]  # Filter based on list of column names
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Step 1: Create a SimpleImputer object with strategy 'most_frequent'
imputer = SimpleImputer(strategy='most_frequent')

# Step 2: Fit the imputer on the 'Cabin' column
imputer.fit(data[['Cabin']])

# Step 3: Transform (fill) the 'Cabin' column and ensure it's a pandas Series
data['Cabin'] = pd.Series(imputer.transform(data[['Cabin']]).ravel(), index=data.index)

# Output the transformed column

# Verify the output
print(data['Cabin'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
survival_counts = data.groupby('Survived')['PassengerId'].count()
plt.figure(figsize=(8, 6))  # Set figure size
survival_counts.plot(kind='bar', color=['red', 'green'])  # Colors for survived/not survived
plt.title("Survival Rates on the Titanic")
plt.xlabel("Survived (0: No, 1: Yes)")
plt.ylabel("Number of Passengers")
plt.xticks([0, 1], ['No', 'Yes'])  # Set custom x-axis labels
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines
for i, v in enumerate(survival_counts):
    plt.text(i, v + 0.1, str(v), ha='center', va='bottom', fontsize=12)

plt.tight_layout()  # Adjust layout for better appearance
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Group data by sex and calculate survival counts (for both survived and not survived)
survival_by_sex = data.groupby('Sex')['Survived'].value_counts().unstack()

# Create the bar chart
plt.figure(figsize=(8, 6))  # Set figure size
survival_by_sex.plot(kind='bar', stacked=False, color=['lightcoral', 'lightblue'])  # Colors for survived/not survived

# Labels, title, and grid
plt.xlabel("Sex")
plt.ylabel("Number of Passengers")
plt.title("Survival Rates by Sex on the Titanic")
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability if needed
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Optional: Show percentages on top of bars (modify as needed)
for col in survival_by_sex.columns:
    for i, v in enumerate(survival_by_sex[col]):
        total = survival_by_sex.sum(axis=1)[i]  # Get total passengers per sex group
        percentage = (v / total) * 100  # Calculate percentage
        plt.text(i, v + 0.1, f"{v} ({percentage:.1f}%)", ha='center', va='bottom', fontsize=12)  # Display count and percentage

plt.legend(title='Survived')  # Add legend

plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder

features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']  # Example features
target = 'Survived'

label_encoder = LabelEncoder()
data['Sex'] = label_encoder.fit_transform(data['Sex'])
data['Embarked'] = label_encoder.fit_transform(data['Embarked'])

X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)

model = LogisticRegression(max_iter=200)  # Increase max_iter if the default 100 is not sufficient
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

conf_matrix = confusion_matrix(y_test, y_pred)  # Use a different variable name to avoid overwriting the function name
print("\nConfusion Matrix:\n", conf_matrix)
```

```python
import matplotlib.pyplot as plt
import numpy as np
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
plt.imshow(conf_matrix, interpolation='nearest', cmap='Blues')
plt.title('Confusion Matrix for Survival Prediction (Heatmap)')
plt.colorbar()
tick_marks = np.arange(len(set(y_test)))
plt.xticks(tick_marks, tick_marks)
plt.yticks(tick_marks, tick_marks)
thresh = conf_matrix.max() / 2.
for i, j in np.ndindex(conf_matrix.shape):
    plt.text(j, i, format(conf_matrix[i, j], 'd'),
             horizontalalignment="center",
             color="white" if conf_matrix[i, j] > thresh else "black")

plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.tight_layout()
plt.show()
```
